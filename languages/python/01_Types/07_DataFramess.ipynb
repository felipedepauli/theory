```python
import numpy as np
import pandas as pd
```

```python
# Vamos importar a função randn que cria números aleatórios como um array com a dimensão que a gente quiser.
from numpy.random import randn
np.random.seed(101)
randn(5, 2)
```

```python
# Vamos definir nosso primeiro dataframe.
# pd.Dataframe( data , index , columns )
df = pd.DataFrame(randn(5, 4), ['A', 'B', 'C', 'D', 'E'], ['W', 'X', 'Y', 'Z'])
print(type(df))      # Coleção de séries.
print(type(df['X'])) # Veja como cada coluna é uma série.
df
# Será uma série, porque por mais que você peça apenas uma coluna, você terá a coluna com o index.
```

```python
# Se você quiser, poderá ter duas colunas na parada. Dá um bizu aí.
df[['W', 'X']]
```

```python
# Perceba que você tem que colocar dois colchotes, senão dará erro.
# df['W', 'X']
```

## Criando novas colunas

```python
# Dá para criar novas colunas bem facilmente em um dataframe.
# Faremos muito isso para poder criar colunas que consolidam informações de outras colunas.
# Vamos criar uma coluna de soma das duas últimas colunas.
print(df)
df['Sum'] = df['Y'] + df['Z']
print(df)
```

## Removendo linhas e colunas

```python
# Antes de remover, vamos inserir mais uma coluna, que vai ser uma copia da coluna W.
df['Copy'] = df['W']
df
```

```python
# O método de remoção está na classe do DataFrame. Então, a remoção será chamada a partir da própria instância.
# df.drop(['Copy'])
# Perceba o erro...
```

```python
# O que aconteceu é que o método rm funciona tanto para remover linhas quanto colunas.
# É necessário definir o que se quer remover. Sem nada for especificado, a função assumirá que ser pretende remover a linha.
print(df)
novo_df = df.drop('E')
print(novo_df)
df
# Viu que loucura? O drop retorna uma cópia sem a linha 'E', sem mexer no dataframe original. Se você quiser mudá-lo, terá que atribuir a ele mesmo.
```

```python
df = df.drop('Copy', axis=1)
df
```

```python
# Mas tem jeito mais fácil de fazer isso aí. Você pode usar o inplace.
df.drop('Sum', axis=1, inplace=True)
df
# Já é!
```

# Comandos Caóticos
### ZIP

```python
# O zip permite juntar duas listas em uma outra coisa.
indexes = ['1', '1', '1', '2', '2', '2']
print(type(indexes))
values  = ["Satanas", "Satanas", "Saci", "John", "John", "Manezinho da Ilha"]
print(type(values))
lista       = list(zip(indexes, values))
dicionario  = dict(zip(indexes, values))
sequencia   = set(zip(indexes, values))
print(type(lista))
print(lista)
print(type(dicionario))
print(dicionario)
print(type(sequencia))
print(sequencia)
```

### pd.MultiIndex.frol_tuples

## Removendo Null

```python
# Cê lembra que tivemos células com Null quando fizemos
df = pd.DataFrame(randn(10, 5), np.arange(0,10), ["a", "b", "c", "d", "e"])
satanas = df[df>-1]
satanas
```

```python
# Qualquer linha ou coluna com NaN
satanas.dropna()
```

```python
# Qualquer coluna com NaN
satanas.dropna(axis=1)
```

```python
# df.dropna(axis=1, thresh=2)  Não ornou
satanas.fillna(value='[--XXX--]')
```

```python
# Sabendo usar a parada, dá para ser malandrão.
satanas['a'].fillna(value=satanas['a'].mean())
```

```python
satanas.describe()
```
